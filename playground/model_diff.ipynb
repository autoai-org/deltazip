```python
# compare the difference between base model and the fine-tuned model
from transformers import AutoModel

base_model = AutoModel.from_pretrained('facebook/opt-1.3b')
finetuned_model = AutoModel.from_pretrained('../.cache/models/opt-1.3b/word_semantics/')
```

```python
diff = {}
origin = {}
for name1, param1 in base_model.named_parameters():
    diff[name1] = finetuned_model.state_dict()[name1] - param1
    origin[name1] = param1
```

```python
for key in diff.keys():
    relative_err = diff[key] - origin[key]
    print(key, relative_err.abs().mean().item())
```
